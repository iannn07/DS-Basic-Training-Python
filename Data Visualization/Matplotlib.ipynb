# Data Visualization using Matplotlib

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
# Basic Example
fig = plt.figure()
axis = plt.axes()
plt.show()
```

### Lineplot

```python
import numpy as np
```

```python
# Data Initialization

# By default, linspace num index is start from 1000 or 1 in the plot
x = np.linspace(50, 100)
y = np.sin(x)
```

```python
fig = plt.figure()
axis = plt.axes()

# Plot initialization
axis.plot(x, y)
plt.show()
```

```python
# We also may use this method and combine the np into the plot() function

plt.plot(x, np.cos(y))
# And to save the file, we may use this
plt.savefig('Lineplot.png')
plt.show()

# Make sure that when we want to save the fig, it's above the plt.show()
# plt.savefig('Test.png')
```

#### Labels & Titles

```python
plt.xlabel('x')
plt.ylabel('y')
plt.title('Function of cos(x)')
plt.plot(x, np.cos(x))
plt.show()
```

```python
# We also could to limit the axis by using .xlim and .ylim like this
plt.xlabel('x')
plt.ylabel('y')
plt.title('Function of tan(x)')
plt.xlim(75, 100)
plt.ylim(-25, 10)
plt.plot(x, np.tan(x))
plt.show()
```

### Multiple Lines

We may combine two line using plt.plot() like this

```python
plt.plot(x, np.cos(x))
plt.plot(x, np.sin(y))

plt.xlabel('x')
plt.ylabel('y')

# We also could add the legend by doing this
plt.legend(['cos(x)', 'sin(y)'])
# This legend will show the line color respectively and sequentially

plt.xlim(50, 60)

plt.show()
```

```python
# Or we may use this method to display the legend

plt.plot(x, np.cos(x), label='cos(x)')
plt.plot(x, np.sin(y), label='sin(y)')

plt.xlabel('x')
plt.ylabel('y')
plt.xlim(50, 60)

plt.legend()
plt.show()
```
