# Plotting

This includes scatter, and others

```python
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('ggplot')
```

```python
# Let's take the data from Sololearn
us_president = pd.read_csv('https://sololearn.com/uploads/files/president_heights_party.csv', index_col='name')
```

```python
us_president.columns
```

```python
first_five = us_president[:5]
print(first_five)

plt.scatter(first_five['age'], first_five['height'])

plt.xlabel('Age')
plt.ylabel('Height')
plt.savefig('Scatter Plot.png')
plt.show()
```

## Plotting w/ Pandas

```python
import pandas as pd
```

```python
us_president.plot(kind='scatter', x='age', y='height', title='US President')
```

## Histogram

Histogram emphasizes the continuous data that are measured on an interval scale

```python
# bins is the the count of bar towers in histogram
# Let's try it using pandas
us_president['age'].plot(kind='hist', title='height', bins=5)
```

```python
plt.hist(us_president['age'], bins=5)

# See? It returns the same result by using Matplotlib
```

## Boxplot

This plot is used to show the distributions of numeric data values, especially when we want to compare our data with multiple groups. It's very useful when we want to describe each individual statistic. 

The orange box indicates the interquartile range (between the first and the third quartile). Meanwhile the blue line inside represent the median of the data and the black line represents the maximum and the minimum of the dataset.

```python
us_president.boxplot(column='age')
```

```python
plt.boxplot(us_president['age'])
```

## Barplot

It will show us the distribution of data over several groups/category

```python
president_party = us_president['party'].value_counts()
president_party
```

```python
president_party.plot(kind='bar')
```
